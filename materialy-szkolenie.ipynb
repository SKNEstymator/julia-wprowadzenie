Wygenerujemy sobie następujący zestaw danych:

$$
\begin{cases}
X_1 & \sim N(0, 5) \\
X_2 & \sim \chi^2(2) \\
\epsilon & \sim N(0,1) \\
Y &= 2*X_1 + 4*X_2 + \epsilon
\end{cases}
$$

```julia
Random.seed!(123) ## bang symbol -- nadpisanie istniejącego obietu
N = 10_000 ## 10000
X₁ = rand(Normal(0,5), N)
X₂ = rand(Chisq(2), N)
ϵ = rand(Normal(), N)
Y = 2 .* X₁ .+ 4 .* X₂ .+ ϵ 
## równoważnie
Y = @. 2*X₁ + 4*X₂ + ϵ
```

```julia
√Y[1]
```

```julia
histogram(Y, legend = false, color = "white")
```

```julia
scatter(X₁, Y, color = "white", legend = false, markershape = :circle)
```
